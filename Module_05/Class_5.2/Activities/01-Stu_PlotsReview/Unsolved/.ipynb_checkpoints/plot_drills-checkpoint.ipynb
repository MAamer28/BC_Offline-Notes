```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(0, len(gyms))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

plt.title("NYC Gym Popularity")
plt.xlabel("Gym Name")
plt.ylabel("Number of Members")

# Finish the plot

plt.bar(gyms, members, color = ["red", "blue", "green", "orange"], alpha = 0.75)
plt.show()
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.title("Sin from 0 to 2$\pi$")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Finish the plot
plt.plot(x_axis, sin, linewidth = 0.5, marker = "o", color = "blue")
plt.hlines(0, 0, 7, alpha = 0.25)
plt.show()
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.title("NYC Gym Popularity")

# Finish the plot
plt.pie(members, explode = explode, labels = gyms, colors = colors, 
        autopct = "%1.1f%%", shadow = True, startangle = 0)
plt.axis("equal")
plt.show()
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.title("Running Time of FakeSort for Sample Input Sizes")
plt.xlabel("Length of Input Array")
plt.ylabel("Time to Sort (s)")

# Finish the plot
plt.scatter(x_axis, times, marker = "o", facecolors = "red", edgecolors = "black",
           alpha = 0.75)
```
