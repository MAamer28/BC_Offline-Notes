```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"]
cars_in_cities = [214.7, 564.4, 416.5, 466.7, 350.6]
x_axis = np.arange(len(cars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(cities, cars_in_cities, color = "b", alpha = 0.5)
plt.show()
```

```python
# Create the ticks for our bar chart's x axis
```

```python
# Set the limits of the x axis
```

```python
# Set the limits of the y axis
plt.ylim(0, max(cars_in_cities)+50)
plt.show()
```

```python
# Give the chart a title, x label, and y label
plt.title("Car Useage in Five US Cities")
plt.xlabel("City")
plt.ylabel("Cars per 1,000 Residents 16+")
plt.show()
```
