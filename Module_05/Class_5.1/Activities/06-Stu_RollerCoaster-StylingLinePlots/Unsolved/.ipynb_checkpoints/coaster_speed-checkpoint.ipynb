```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# Plot the charts and apply some styling
dangerdrop_handle, = plt.plot(time, danger_drop_speeds, color = "red", label = "Danger Drop")
railgun_handle, = plt.plot(time, railgun_speeds, color = "blue", label = "RailGun")
```

```python
# Add labels to X and Y axes :: Add title
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.title("Coaster Speed Over Time")
plt.show()
```

```python
# Set the limits for the X and Y axes
plt.xlim(0, 120)
plt.ylim(0, 100)
plt.show()
```

```python
# Create a legend for the chart
plt.legend()
plt.show()
```

```python
# Add in a grid for the chart
plt.grid()
plt.show()
```
