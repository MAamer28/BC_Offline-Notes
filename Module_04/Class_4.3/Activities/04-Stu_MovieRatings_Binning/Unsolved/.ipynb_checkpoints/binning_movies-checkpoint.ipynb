```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
filepath = Path("Resources/movie_scores.csv")

movies_df = pd.read_csv(filepath)
movies_df.head()
```

```python
# Figure out the minimum and maximum IMDB user vote count
raters_df = movies_df.sort_values("IMDB_user_vote_count", ascending=False)
raters_df.head()
```

```python
# Create bins in which to place values based upon IMDB vote count
bins = [0, 1000, 10000, 25000, 50000, 100000, 150000, 200000, 250000, 300000, 350000]
# Create labels for these bins
bin_labels = ["Under 1,000 Voters", "1,000 to 10,000 Votes", "10K to 25K Votes", "25K to 50K Votes", "50K to 100K Votes", 
              "100K to 150K Votes", "150K to 200K Votes", "200K to 250K Votes", "250K to 300K Votes", "Greater than 300K Votes"]
```

```python
# Slice the data and place it into bins
raters_df["IMDB Popularity"] = pd.cut(raters_df["IMDB_user_vote_count"], bins, labels=bin_labels)
```

```python
# Place the data series into a new column inside of the DataFrame
raters_df.head()
```

```python
# Create a GroupBy object based upon "IMDB User Votes Group"
IMDB_df = raters_df.groupby("IMDB Popularity")

print(IMDB_df["IMDB Popularity"].count())

# Find how many rows fall into each bin


# Get the average of each of the first 5 rating columns within the GroupBy object
IMDB_df[["RottenTomatoes", "RottenTomatoes_User", "Metacritic", "Metacritic_User", "IMDB"]].mean()
```
