```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepaths into variable
avg_file = Path("Resources/state_avg.csv")
total_file = Path("Resources/state_totals.csv")
```

```python
# Read in files.
stateavg_df = pd.read_csv(avg_file)
statetotals_df = pd.read_csv(total_file)
```

```python
# Display the state averages.
stateavg_df.head()
```

```python
# Display the state totals.
statetotals_df.head()
```

```python
# Merge the two DataFrames together based on the Year and State they share
merged_df = pd.merge(stateavg_df, statetotals_df, on=["Year", "State"])
merged_df.head()
```

```python
# Create a DataFrame that filters the data on only 2019
merged2019_df = merged_df.loc[merged_df["Year"] == 2019]
merged2019_df.head()
```

```python
# Add a new column that calculates the Poverty Rate
merged2019_df["Poverty Rate (%)"] = (merged_df["Total Population in Poverty"] / merged_df["Total Population"])*100
merged2019_df.head()
```
